```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common import NoSuchElementException, TimeoutException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#espera
import time
chrome = webdriver.Chrome()
```

```python
# Exemplo 05
# https://curso-web-scraping.pages.dev/#/exemplo/5
# server local
pagina = 'http://localhost:8000/#/exemplo/5'
chrome.get(pagina)
```

```python
# Configurar Espera de elementos da pagina
tempo_max = 35  # tempo maximo de espera
tempo_tentativa = 1  # a cada segundo
wait = WebDriverWait(driver=chrome, timeout=tempo_max, poll_frequency=tempo_tentativa)
```

```python
# Eventos na pagina
# tittle_is
chrome.refresh()
%time wait.until(EC.title_is(title='Programador'))  # % time do jupyter notebook
```

```python
# Eventos na pagina
# title_contains
chrome.refresh()
%time wait.until(EC.title_contains(title='amado'))  # % time do jupyter notebook
```

```python
# Eventos na pagina
# presence_of_element_located
chrome.refresh()
%time wait.until(EC.presence_of_element_located(locator=(By.ID, 'star1')))  # % time do jupyter notebook
```

```python
# Eventos na pagina
# presence_of_all_elements_located
chrome.refresh()
%time wait.until(EC.presence_of_all_elements_located(locator=(By.NAME, 'stars')))  # % time do jupyter notebook - lista de estrelas
```

```python
# Eventos na pagina
# visibility_of_element_located
chrome.refresh()
%time wait.until(EC.presence_of_element_located(locator=(By.ID, 'star3')))  # presente no HTML (Sim)
%time wait.until(EC.visibility_of_element_located(locator=(By.ID, 'star3')))  # % time do jupyter notebook - estrelas nao visivel
```

```python
# element_to_be_clickable
# button#telegram > a
chrome.refresh()
%time wait.until(EC.element_to_be_clickable(mark=(By.CSS_SELECTOR, 'button#telegram')))  # % time do jupyter notebook - clicavel na tela
```

```python
# staleness_of
chrome.refresh()
ele_some = chrome.find_element(By.ID, 'star4')
%time wait.until(EC.staleness_of(element=ele_some))   # verifica se elemento sumiu 
```

```python
# element_attribute_to_include
%time
chrome.refresh()
wait.until(
    EC.element_attribute_to_include(
        locator=(By.ID, 'star5'),
        attribute_='name'
))   # Aguarda elemento ganhar atributo
```

```python
# Popup
# alert_is_present
%time
chrome.refresh()
alerta = wait.until(EC.alert_is_present())
print(type(alerta))
print(alerta.text)

time.sleep(2)
# Clicar em Ok.
alerta.accept()      
# Cancelar Alerta
# alerta.dismiss()  
```
