```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common import NoSuchElementException, TimeoutException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Rendereizar HTML no Notebook
from IPython.display import HTML

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/6
# server local
pagina = 'http://localhost:8000/#/exemplo/6'
chrome.get(pagina)
chrome.implicitly_wait(time_to_wait=10)  # Aguarda 10 seg antes de levantar exceção.
```

```python
estados = chrome.find_element(By.ID, 'estados')
estados
```

Atributos

```python
estados.get_attribute('class')
```

```python
estados.get_attribute('id')
```

Propriedade (DOM)

```python
estados.get_property('className')
```

```python
# estados > div
filhos = estados.get_property('children')
for filho in filhos:
    # print ('Filho:', filho.text)
    netos = filho.get_property('children')
    for neto in netos:
        if neto.tag_name == 'label':
            print('Neto: ', neto.text, ' (label).')  # #estados > div:nth-child(1) > label
        elif neto.tag_name == 'div':
            print('\t -', neto.text, ' (div).')
```
