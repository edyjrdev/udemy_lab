```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# Filtrar por Classe
# CLASS_NAME
lista_input = chrome.find_elements(By.CLASS_NAME, 'styled-input')
lista_input
```

```python
for item in lista_input:
    print(item.get_property('value'))
```

```python
# Com duas ou mais classes: styled-input e optional-info. (separar com .)
lista_classes = chrome.find_elements(By.CLASS_NAME, 'styled-input.optional-info')
lista_classes
```

```python
for ele in lista_classes:
    print(ele.get_property('value'))
```
