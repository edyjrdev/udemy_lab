```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common import NoSuchElementException, TimeoutException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chrome = webdriver.Chrome()

import os, time
import base64  # para manipular base64 e salvar em PDF
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
chrome.implicitly_wait(time_to_wait=10)  # aguardar 10 seg antes de levantar exceção
```

```python
# criar pasta
pasta_screen = './captura_tela'

if not os.path.exists(pasta_screen):
    os.mkdir(pasta_screen)
```

ScreenShot da Pagina

```python
for i in range(5):
    arquivo = f'captura_tela {i}.png'
    nome_arquivo = os.path.join(pasta_screen, arquivo)
    chrome.save_screenshot(filename=nome_arquivo)
    chrome.refresh()
    time.sleep(1)
```

```python
for i in range(11,16):
    arquivo = f'captura_tela {i}.png'
    nome_arquivo = os.path.join(pasta_screen, arquivo)
    chrome.get_screenshot_as_file(filename=nome_arquivo)
    chrome.refresh()
    time.sleep(1)
```

Arquivo PNG (Binário)

```python
# Criar Arquivo no OS
arquivo = f'captura_tela_bin.png'
nome_arquivo = os.path.join(pasta_screen, arquivo)
with open(file=nome_arquivo, mode='wb') as file_png:
    data = chrome.get_screenshot_as_png()  # captura png binário
    file_png.write(data) # salva conteudo no arquivo
```

ScreenShoot de Elemento

```python
arquivo = 'div_main.png'
nome_arquivo = os.path.join(pasta_screen, arquivo)
div_main = chrome.find_element(By.ID, 'main')
div_main.screenshot(filename=nome_arquivo)

"""
//*[@id="main"]/main/div[2]
"""
arquivo = 'form.png'
nome_arquivo = os.path.join(pasta_screen, arquivo)
form = chrome.find_element(By.XPATH, '//*[@id="main"]/main/div[2]')
form.get_attribute('outerHTML')
form.screenshot(filename=nome_arquivo)

arquivo = 'redes.png'
nome_arquivo = os.path.join(pasta_screen, arquivo)
social = chrome.find_element(By.ID, 'social')
social.get_attribute('outerHTML')
social.screenshot(filename=nome_arquivo)

```

Salvar PDF

```python
# Criar Arquivo no OS
arquivo = f'tela_impressa.pdf'
nome_arquivo = os.path.join(pasta_screen, arquivo)

"""
<button data-drawer-target="default-sidebar" data-drawer-toggle="default-sidebar" aria-controls="default-sidebar" type="button" class="fixed z-50 top-0 left-0 items-center p-2 mt-2 ml-3 text-sm text-gray-500 rounded-lg hover:bg-gray-100 focus:outline-none focus:ring-2 focus:ring-gray-200 dark:text-gray-400 dark:hover:bg-gray-700 dark:focus:ring-gray-600">
<span class="sr-only">Abrir menu</span><svg class="w-6 h-6" aria-hidden="true" fill="currentColor" viewBox="0 0 20 20" xmlns:xlink="http://www.w3.org/2000/svg"><path clip-rule="evenodd" fill-rule="evenodd" d="M2 4.75A.75.75 0 012.75 4h14.5a.75.75 0 010 1.5H2.75A.75.75 0 012 4.75zm0 10.5a.75.75 0 01.75-.75h7.5a.75.75 0 010 1.5h-7.5a.75.75 0 01-.75-.75zM2 10a.75.75 0 01.75-.75h14.5a.75.75 0 010 1.5H2.75A.75.75 0 012 10z"></path></svg></button>

//*[@id="root"]/div/div[1]/button

"""

# btn_menu = chrome.find_element(By.CSS_SELECTOR, 'button > span')
btn_menu = chrome.find_element(By.XPATH, '//*[@id="root"]/div/div[1]/button')

print(btn_menu.get_property('innerText'))
btn_menu.click()  # recolher menu
time.sleep(1)

with open(file=nome_arquivo, mode='wb') as file_pdf:
    chrome.maximize_window()
    data64 = chrome.print_page()  # captura impressao em base64
    data_binario = base64.b64decode(data64)
    file_pdf.write(data_binario) # salva conteudo no arquivo
```
