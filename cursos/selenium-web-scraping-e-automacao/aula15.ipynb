```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common import NoSuchElementException, TimeoutException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#
import urllib.request
import os
chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/7
# server local
pagina = 'http://localhost:8000/#/exemplo/7'
chrome.get(pagina)
chrome.implicitly_wait(time_to_wait=10)  # aguardar 10 seg antes de levantar exceção
```

```python
galeria = chrome.find_element(By.ID, 'galeria')
imagens = galeria.find_elements(By.ID, 'img') 
```

```python
pasta_imagens = './download'

if not os.path.exists(pasta_imagens):
    os.mkdir(pasta_imagens)
```

```python
# direto
imagens = chrome.find_elements(By.CSS_SELECTOR, '#galeria img')

for imagem in imagens:
    # pegar url da imagem = src
    url = imagem.get_attribute('src')
    id = imagem.get_property('id')
    print(id, '-', url)
    
    nome_arquivo = id + '.jpg'
    arquivo_path = os.path.join(pasta_imagens, nome_arquivo)
    print('baixando ->', arquivo_path)

    # baixar imagem
    
    urllib.request.urlretrieve(url,arquivo_path)

    
    
```
