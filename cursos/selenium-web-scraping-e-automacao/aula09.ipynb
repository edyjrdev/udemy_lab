```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

Atributo e Propriedade

```python
# Atributo vem do HTML (TAGs)
atri = chrome.find_element(By.ID, 'zodiac').get_attribute('id')
atri
```

```python
# Propriedade vem do DOM (JavaScript) - nao pode sobrescrever palavra reservada
prop = chrome.find_element(By.ID, 'zodiac').get_property('id')
prop
```

```python
# class
atri = chrome.find_element(By.ID, 'zodiac').get_attribute('class')
prop = chrome.find_element(By.ID, 'zodiac').get_property('className') # evitar sobrescrever class
print( atri,'\n' , prop)
```

```python
# data-optional
atri = chrome.find_element(By.ID, 'zodiac').get_attribute('data-optional')
prop = chrome.find_element(By.ID, 'zodiac').get_property('dataset') # encapsula em dicionario optional
print( atri,'\n' , prop['optional'])
```
