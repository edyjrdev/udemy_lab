```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common import NoSuchElementException, TimeoutException, InvalidSessionIdException, UnexpectedAlertPresentException, NoAlertPresentException
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
chrome.implicitly_wait(time_to_wait=10)  # aguardar 10 seg antes de levantar exceção
```

Criar Nova Janela

```python
for n in range(2,8):
    chrome.switch_to.new_window("tab")  # abrir como aba
    chrome.get(f'http://localhost:8000/#/exemplo/{n}')

chrome.switch_to.new_window(type_hint='window')  # abrir como janela
chrome.get('https://www.bb.com.br')
```

Controlando Abas do Navegador

```python
lista_abas = chrome.window_handles # incluindo guia do inspetor se estiver aberto e janelas abertas no navegador  
lista_abas
```

Dicionario de Janelas

```python
def cria_dicionario_janelas():
    dic_janelas = {}

    chrome.switch_to.window(window_name=chrome.window_handles[0]) # pega ultima janela aberta e torna ativa  # ativa janela

    for aba in chrome.window_handles:
        chrome.switch_to.window(window_name=aba)
        titulo = chrome.title
        print(titulo, ":", aba, "url:", chrome.current_url)
        dic_janelas[titulo] = aba
            
    return dic_janelas
```

Trocar de Aba/Janela

```python
janelas = cria_dicionario_janelas()

# criar nova janela a partir do dicionario de janelas

for chave, valor in janelas.items():
    chrome.switch_to.window(window_name=valor)
    print(chave, "➡️", chrome.current_url)
```

Fechar

```python
# fechar ultima janela aberta
try:
    print('Fechar ', chrome.title)
    chrome.switch_to.window(window_name=chrome.window_handles[-1])
    chrome.close()  # fica sem janela ativa.
    chrome.switch_to.window(window_name=chrome.window_handles[-1]) # pega ultima janela aberta e torna ativa
    print('Janela ativa', chrome.title)

except InvalidSessionIdException:
    print('O chrome foi fechado.')
except Exception as e:
    print(e.msg)
```

Alertas

```python
time.sleep(3)

# ativa janela
chrome.switch_to.window(window_name=chrome.window_handles[-1]) # pega ultima janela aberta e torna ativa
print('Janela ativa', chrome.title)

# chrome.switch_to.new_window(type_hint='window')  # abrir como janela
# chrome.get(f'http://localhost:8000/#/exemplo/5')

# pega janelas abertas
janelas = cria_dicionario_janelas()
titulos = janelas.keys()

tilulo_exemplo_5_ini = 'Exemplo 5'
tilulo_exemplo_5_din = 'Programador Aventureiro'

# ir para exemplo 5 para ativar alerta - Titulo Dinamico: Exemplo 5 => Programador Aventureiro
if tilulo_exemplo_5_ini in titulos:
    print("Achei", tilulo_exemplo_5_ini)
    chrome.switch_to.window(window_name=janelas[tilulo_exemplo_5_ini])
elif tilulo_exemplo_5_din in titulos: 
    print("Achei", tilulo_exemplo_5_din)
    chrome.switch_to.window(window_name=janelas[tilulo_exemplo_5_din])
else:
    raise Exception('Exemplo 5 fechada.')

btn_alert = chrome.find_element(By.CLASS_NAME, 'btn-primary')
print('Clicar em ', btn_alert.text)
btn_alert.click()  #Alerta ativo


try:
    janela_alerta = chrome.switch_to.alert
    print(janela_alerta.text)
    janela_alerta.accept()  # Clicar em Ok
    # janela_alerta.dismiss()  # Fechar Alerta   
    print('Alerta Fechado')

    # volta para ativar janela
    chrome.switch_to.window(window_name=chrome.window_handles[0])
    
except NoAlertPresentException:
    print('Alerta não detectado em Janela')
except Exception as e:
    print(e.msg)
```
