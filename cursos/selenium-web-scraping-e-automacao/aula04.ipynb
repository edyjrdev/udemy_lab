```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# LINK_TEXT
try:
    chrome.find_elements(By.LINK_TEXT, 'Instagram')
except Exception as e:
    print(f'Ocorreu um erro:{e}')
    
```

```python
# PARCIAL_LINK_TEXT
elementos = chrome.find_elements(By.PARTIAL_LINK_TEXT, 'gram')
for e in elementos:
    print(e.text)
```
