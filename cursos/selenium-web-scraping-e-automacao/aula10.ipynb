```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 02
# https://curso-web-scraping.pages.dev/#/exemplo/2
# server local
pagina = 'http://localhost:8000/#/exemplo/2'
chrome.get(pagina)
```

Fazer Login

```python
# mapeando campos
in_email = chrome.find_element(By.NAME, 'email')
in_senha = chrome.find_element(By.NAME, 'senha')
bt_enviar = chrome.find_element(By.CSS_SELECTOR, 'button.btn-primary')
```

```python
# limpar formulario
in_email.clear()
in_senha.clear()

# preencher campos
in_email.send_keys('edyjrdev@gmail.com')
in_senha.send_keys('senhaSecreta')

bt_enviar.click()  # submit do form tem nos elementos # in_senha.submit()
```
