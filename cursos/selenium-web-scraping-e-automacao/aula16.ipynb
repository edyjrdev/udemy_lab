```python
# cria classe com dataclass

@dataclass
class Usuario:
    avatar: str
    nome: str
    profissao: str
    email: str
    perfil: str
    estado: str
```

```python
registros = []

for linha in linhas:  # tr = linha
    colunas = linha.get_property('children')  # th = coluna
    i = 0
    for coluna in colunas:
        print(str(i),'=', coluna.text, '|', end='')
        i+=1

    print('\n')

    usuario = Usuario(
        avatar=colunas[0].find_element(By.TAG_NAME, 'img').get_property('src'),
        nome=colunas[1].text,
        profissao=colunas[2].text,
        email=colunas[3].text,
        perfil=colunas[4].text,
        estado=colunas[5].text
    )
    registros.append(usuario)

for i, user in enumerate(registros, start=1):
    print(i, user.email)
```

```python
def raspar_dados_tabela():
    registros = []
    linhas = chrome.find_elements(By.CSS_SELECTOR, '#tabela-usuarios tbody tr')
    
    for linha in linhas:  # tr = linha
        colunas = linha.get_property('children')  # th = coluna
        
        usuario = Usuario(
            avatar=colunas[0].find_element(By.TAG_NAME, 'img').get_property('src'),
            nome=colunas[1].text,
            profissao=colunas[2].text,
            email=colunas[3].text,
            perfil=colunas[4].text,
            estado=colunas[5].text
        )
        registros.append(usuario)
    
    return registros
```

```python
raspar_dados_tabela()
```

```python
lista_usuarios = []  # lista de usuarios do site
total_pagina = 0
# Raspagem da Paginacao até condição de saída.
while True:
    total_pagina +=1
    dados_pagina = raspar_dados_tabela()  # Raspa pagina atual
    lista_usuarios.extend(dados_pagina) # inclui lista da pagina na lista do site

    #Se não conseguir clicar em Próxima - Chegou na Ultima Pagina
    try:
        proximo = chrome.find_element(By.XPATH, '//main//button[contains(text(), "Próxima")]')
        proximo.click()  # Na ultima pagina dará erro
    except Exception as e:
        print('Não tem próxima')
        break  # Paralisa raspagem

print('Paginas=', total_pagina, ' Usuarios=', len(lista_usuarios))
```
