```python
from selenium import webdriver
from selenium.webdriver.common.by import By
# Aguardar Renderização
from selenium.webdriver.support.wait import WebDriverWait   
from selenium.webdriver.support import expected_conditions as EC  # Eventos (Condições) Esperados
from selenium.common.exceptions import NoSuchElementException, TimeoutException
chrome = webdriver.Chrome()
```

```python
# Exemplo 04
# https://curso-web-scraping.pages.dev/#/exemplo/4
# server local
pagina = 'http://localhost:8000/#/exemplo/4'
chrome.get(pagina)
```

```python
# CONFIG
timeout_default = 2  # pagina configurada com 3 segs de delay
check_in_sec = 1  # frequencia de teste
timeout_geral = 3
```

```python
# inspecionar eventos
dir(EC)
```

```python
chrome.refresh()
# tentar achar elemento
# div vantagens
try:
    div = chrome.find_element(By.ID, 'vantagens')  # NoSuchElementException chegou antes de renderizar
except NoSuchElementException as err:
    print(f'Não achou elemento.{err.msg}')
except Exception as err:
    print(f'Erro Desconhecido.{err.msg}')
```

```python
chrome.refresh()  #reiniciar carregamento da pagina
# Configurar Timeout
# Configurar Tempo de Checagem
wait = WebDriverWait(driver=chrome, timeout=timeout_default, poll_frequency=check_in_sec)
# Locator = find_element
try:
    ele = wait.until(
        method=EC.presence_of_element_located(locator=(By.ID, 'vantagens')),
        message='ERR: Div Vantagens não localizada.')
    print(ele.get_attribute('id'))
except TimeoutException as err:
    print(f'Pagina lenta.{err.msg}')  
```

Alternativa
tempo implicito, não sabe qual evento ocorrerá

```python
chrome.implicitly_wait(time_to_wait=timeout_geral)  # aguarda tempo geral, sem conectar a evento (EC)
div = chrome.find_element(By.ID, 'vantagens')
div
```
