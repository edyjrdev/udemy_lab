```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# Selecionar por Name
nome_completo = chrome.find_element(By.NAME, 'fullname').get_property('value')
profissao = chrome.find_element(By.NAME, 'role').get_property('value')
signo = chrome.find_element(By.NAME, 'zodiacSign').get_property('value')
genero = chrome.find_element(By.NAME, 'genderOfBirth').get_property('value')

print(f"""
{nome_completo=}
{profissao=}
{signo=}
{genero=}

""")
```
