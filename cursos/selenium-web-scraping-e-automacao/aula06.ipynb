```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# CSS_SELECTOR
lista_elemento = chrome.find_elements(By.CSS_SELECTOR, 'div#social')
lista_elemento
```

```python
lista_classe = chrome.find_elements(By.CSS_SELECTOR, 'input.styled-input.optional-info')
lista_classe
```

```python
# descendentes de div.main-container que seja input
lista_descendentes = chrome.find_elements(By.CSS_SELECTOR, 'div.main-container input.optional-info')
lista_descendentes
```

```python
# descendentes diretos
lista_links = chrome.find_elements(By.CSS_SELECTOR, 'div#social > a')
lista_links
```

```python
# seletor universal * (hierarquia recursiva)
lista_all_descendentes = chrome.find_elements(By.CSS_SELECTOR, 'div.main-container *')
lista_all_descendentes

for ele in lista_all_descendentes:
    print(ele.tag_name, '=>', ele.get_attribute('class') ) # atributo html
    #print(ele.get_property('className')) # propriedade do DOM (JavaScript)
```

```python
# descentes da div social
lista_social = chrome.find_elements(By.CSS_SELECTOR, 'div#social > *')
lista_social

for ele in lista_social:
    print(ele.tag_name, '=>', ele.get_property('className'))
```

Propriedades e Atributos


```python
elementos = chrome.find_elements(By.CSS_SELECTOR, 'input[name="fullname"]')
elementos
```

```python
linsk = chrome.find_elements(By.CSS_SELECTOR, 'a[href="#"]')
linsk
```

Sufixo

```python
# valor terminado em *ino
inos = chrome.find_elements(By.CSS_SELECTOR, 'input[value$="ino"]')
inos
```

```python
grams = chrome.find_elements(By.CSS_SELECTOR, '[class$="gram"')
grams
for ele in grams:
    print(ele.get_attribute('class'))
```

Prefixo

```python
prefixs = chrome.find_elements(By.CSS_SELECTOR, '[name^="user"]')
prefixs

for ele in prefixs:
    print(ele.get_attribute('name'))
```

|= seleção "algo-"

```python
hifens = chrome.find_elements(By.CSS_SELECTOR, '[class|="text"]')
hifens

for ele in hifens:
    print(ele.tag_name," => ",ele.get_attribute('class'))
```

Substring *=

```python
inputs = chrome.find_elements(By.CSS_SELECTOR, 'input[id*="e"]')

for ele in inputs:
    print(ele.tag_name," => ",ele.get_attribute('id'))
```
