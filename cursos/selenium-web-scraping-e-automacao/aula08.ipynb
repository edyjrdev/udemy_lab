```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# Copiar Seletor no navegador
chrome.find_elements(By.CSS_SELECTOR, '#social > a:nth-child(3)')
```

```python
# Copiar XPath no navegador
# Relativo
chrome.find_elements(By.XPATH, '//*[@id="social"]/a[3]')
# Completo = /html/body/div/div/div[2]/main/div[2]/div/div[2]/a[3]
```
