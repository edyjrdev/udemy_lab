```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import json
import datetime  # Tratar Data Str

chrome = webdriver.Chrome()
```

```python
# Desafio 01
# https://curso-web-scraping.pages.dev/#/desafio/1
# server local
pagina = 'http://localhost:8000/#/desafio/1'
chrome.get(pagina)
```

Ler arquivo desafio_1.json
para cada registro preencher formulario e enviar

```python
# Abre o arquivo para leitura
with open('./desafio_1.json', 'r', encoding='utf-8') as arquivo:
    # Converte o conteúdo JSON em um objeto Python (dicionário/lista)
    lista_dados = json.load(arquivo)

# Exemplo de acesso aos dados
for registro in lista_dados:
    email = registro['email']
    senha =  registro['senha']
    data_nascimento = registro['data-de-nascimento']
    liga_newsletter = bool(registro['newsletter'])
    data_nasc = datetime.datetime.strptime(data_nascimento, '%Y-%m-%d')
    print (email, senha, data_nasc, liga_newsletter, sep='|')

    # Mapear Formulario
    in_email = chrome.find_element(By.NAME, 'email')
    in_senha = chrome.find_element(By.NAME, 'senha')
    sel_dia = Select(chrome.find_element(By.NAME, 'dia'))
    sel_mes = Select(chrome.find_element(By.NAME, 'mes'))
    sel_ano = Select(chrome.find_element(By.NAME, 'ano'))
    bt_check = chrome.find_element(By.CSS_SELECTOR, 'button#airplane-mode')

    bt_check_on = True if bt_check.get_attribute('aria-checked') == 'true' else False


    print(bt_check.get_property('value'))
    # Limpar Formulario
    in_email.clear()
    in_senha.clear()
    # valor default
    sel_dia.select_by_index(0)
    sel_mes.select_by_index(0)
    sel_ano.select_by_index(0)
    
    # Preencher Formulario
    in_email.send_keys(email)
    in_senha.send_keys(senha)
    sel_dia.select_by_visible_text(str(data_nasc.day))
    sel_mes.select_by_visible_text(str(data_nasc.month))
    sel_ano.select_by_visible_text(str(data_nasc.year))

    # ligar newsletter
    print(f'ligar:{liga_newsletter} e check:{bt_check_on}')
    if liga_newsletter != bt_check_on:
        bt_check.click()

    # Enviar
    in_email.submit()
```
