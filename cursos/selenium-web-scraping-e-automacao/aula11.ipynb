```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# trabalhar com combo e list box
from selenium.webdriver.support.ui import Select

chrome = webdriver.Chrome()
```

```python
# Exemplo 03
# https://curso-web-scraping.pages.dev/#/exemplo/3
# server local
pagina = 'http://localhost:8000/#/exemplo/3'
chrome.get(pagina)
```

```python
# Select
regioes = Select(chrome.find_element(By.NAME, 'regiao'))

for regiao in regioes.options:
    print(regiao.get_attribute('value'), '-', regiao.text)
```

```python
# Selecionar Regiao por indice
regioes.select_by_index(3)
```

```python
regioes.select_by_visible_text('Goiás')
```

```python
regioes.select_by_value('2')
```

```python
# opção selecionada no momento
regioes.first_selected_option.text
```

Seleção Multipla

```python
lista_centro_oeste = Select(chrome.find_element(By.CSS_SELECTOR, 'div#multi-select select'))

for regiao in lista_centro_oeste.options:
    print(regiao.get_attribute('value'), '-', regiao.text)
```

```python
lista_centro_oeste.select_by_index(1)
lista_centro_oeste.select_by_index(3)
```

```python
for selecionado in lista_centro_oeste.all_selected_options:
    print(selecionado.get_attribute('value'), '-', selecionado.text)
```

```python
# limpar seleção
lista_centro_oeste.deselect_all()
```

```python
lista_centro_oeste.select_by_visible_text('Distrito Federal')
lista_centro_oeste.select_by_visible_text('Mato Grosso do Sul')
```

```python
lista_centro_oeste.deselect_by_visible_text('Mato Grosso do Sul')
lista_centro_oeste.deselect_by_value('0')
```

```python
# selecionar tudo
for regiao in lista_centro_oeste.options:
    lista_centro_oeste.select_by_visible_text(regiao.text)
```

```python
lista_centro_oeste.deselect_all()
```

Selecionar 2 elemento dos selects

```python
selecoes = chrome.find_elements(By.TAG_NAME, 'select')

for item in selecoes:
    selecao = Select(item)
    selecao.select_by_index(1)  # segundo elemento
```
