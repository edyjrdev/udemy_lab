```python
from selenium import webdriver

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# Classe para Seleção (CSS)
from selenium.webdriver.common.by import By
dir(By)
"""
'CLASS_NAME',
 'CSS_SELECTOR',
 'ID',
 'LINK_TEXT',
 'NAME',
 'PARTIAL_LINK_TEXT',
 'TAG_NAME',
 'XPATH
 """
```

```python
# Propriedade Class do Elemento
user = chrome.find_element(By.ID, 'user')
user.get_property('className')
```

```python
# Formulario
# Nome Completo id=user
nome_completo = user.get_property('value')
# Profissao id=role
role = chrome.find_element(By.ID, 'role')
profissao = role.get_property('value')
# Signo - id=zodiac
zodiac = chrome.find_element(By.ID, 'zodiac')
signo = zodiac.get_property('value')
# Genero - id=gender
gender = chrome.find_element(By.ID, 'gender')
genero = gender.get_property('value')

print(f"""
Dados do formulário:
{nome_completo=} 
{signo=}
{profissao=}
{genero=}.
""")
```
