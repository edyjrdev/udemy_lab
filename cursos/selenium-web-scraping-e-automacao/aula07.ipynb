```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

XPath

```python
# Caminho Absoluto (Completo a partir da root /)
# posicional
chrome.find_elements(By.XPATH, '/html/body/div/div/div[2]')
```

```python
# Por tag
chrome.find_elements(By.XPATH, '//input')
```

Busca por posição

```python
# Por tag
# Posição 1 ou 4
chrome.find_elements(By.XPATH, '//input[1] | //input[4]')
```

Busca por Atributos

```python
# atributos incluir @atributo
chrome.find_elements(By.XPATH, '//input[@id="user"]')
```

```python
# / descentente direto
chrome.find_elements(By.XPATH, '//div[@id="social"]/a[1]')
```

```python
# // descentes gerais abaixo na hierarquia
chrome.find_elements(By.XPATH, '//div[@id="social"]//span')
```
