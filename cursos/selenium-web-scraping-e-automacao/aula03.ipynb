```python
from selenium import webdriver
from selenium.webdriver.common.by import By

chrome = webdriver.Chrome()
```

```python
# Exemplo 01
# https://curso-web-scraping.pages.dev/#/exemplo/1
# server local
pagina = 'http://localhost:8000/#/exemplo/1'
chrome.get(pagina)
```

```python
# TAG_NAME
# chrome.find_element() - retorna primeiro elemento
lista_input = chrome.find_elements(By.TAG_NAME, 'input')
```

```python
for item in lista_input:
    print(item.get_property('value'))
```

```python
nome = lista_input[0].get_property('value')
profissao = lista_input[1].get_property('value')
signo = lista_input[2].get_property('value')
sexo = lista_input[3].get_property('value')

print(f"""
Dados formulario:
{nome=}
{profissao=}
{signo=}
{sexo=}
""")
```

```python
# pegar links das redes sociais em span
lista_spans = chrome.find_elements(By.TAG_NAME, 'span')
```

```python
for span in lista_spans:
    # pegar texto <span>texto</span>
    print(span.get_property('innerText'))
```

```python
# Concatenando buscas
# span dentro da div=social
lista_redes = chrome.find_element(By.ID, 'social').find_elements(By.TAG_NAME, 'span')
```

```python
for rede in lista_redes:
    print(rede.text)
```
